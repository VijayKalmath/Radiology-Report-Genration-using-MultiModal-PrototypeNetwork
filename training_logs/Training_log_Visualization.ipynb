```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
import re
```

```python
log_file = "Training.log"
log_data = {}

with open(log_file) as f : 
    log_lines = f.readlines()
    
for log_line in log_lines :     
    epoch_number = re.findall("\sepoch\s+: (.*)",log_line)
    ce_loss = re.findall("\sce_loss\s+: (.*)",log_line)
    img_constrastive_learning = re.findall("\simg_con\s+: (.*)",log_line)
    text_constrastive_learning = re.findall("\stxt_con\s+: (.*)",log_line)
    img_bce_loss = re.findall("\simg_bce_loss\s+: (.*)",log_line)
    text_bce_loss = re.findall("\stxt_bce_loss\s+: (.*)",log_line)
    val_BLEU_1= re.findall("\sval_BLEU_1\s+: (.*)",log_line)
    val_BLEU_2= re.findall("\sval_BLEU_2\s+: (.*)",log_line)
    val_BLEU_3= re.findall("\sval_BLEU_3\s+: (.*)",log_line)
    val_BLEU_4= re.findall("\sval_BLEU_4\s+: (.*)",log_line)
    val_METEOR= re.findall("\sval_METEOR\s+: (.*)",log_line)
    val_ROUGE_L= re.findall("\sval_ROUGE_L\s+: (.*)",log_line)
    test_BLEU_1= re.findall("\stest_BLEU_1\s+: (.*)",log_line)
    test_BLEU_2= re.findall("\stest_BLEU_2\s+: (.*)",log_line)
    test_BLEU_3= re.findall("\stest_BLEU_3\s+: (.*)",log_line)
    test_BLEU_4= re.findall("\stest_BLEU_4\s+: (.*)",log_line)
    test_METEOR= re.findall("\stest_METEOR\s+: (.*)",log_line)
    test_ROUGE_L  =re.findall("\stest_ROUGE_L\s+: (.*)",log_line)
    
    if epoch_number: 
        EPOCH = int(epoch_number[0])
        log_data[EPOCH] = {} 
    elif ce_loss : 
        log_data[EPOCH]["CE_Loss"] = float(ce_loss[0]) 
    elif img_constrastive_learning : 
        log_data[EPOCH]["img_constrastive_learning"] = float(img_constrastive_learning[0])
    elif text_constrastive_learning : 
        log_data[EPOCH]["text_constrastive_learning"] = float(text_constrastive_learning[0])
    elif img_bce_loss : 
        log_data[EPOCH]["img_bce_loss"] = float(img_bce_loss[0])
    elif text_bce_loss : 
        log_data[EPOCH]["text_bce_loss"] = float(text_bce_loss[0])
    elif val_BLEU_1 : 
        log_data[EPOCH]["val_BLEU_1"] = float(val_BLEU_1[0])
    elif val_BLEU_2 : 
        log_data[EPOCH]["val_BLEU_2"] = float(val_BLEU_2[0])
    elif val_BLEU_3 : 
        log_data[EPOCH]["val_BLEU_3"] = float(val_BLEU_3[0])
    elif val_BLEU_4 : 
        log_data[EPOCH]["val_BLEU_4"] = float(val_BLEU_4[0])
    elif val_METEOR : 
        log_data[EPOCH]["val_METEOR"] = float(val_METEOR[0])
    elif val_ROUGE_L : 
        log_data[EPOCH]["val_ROUGE_L"] = float(val_ROUGE_L[0])
    elif test_BLEU_1 : 
        log_data[EPOCH]["test_BLEU_1"] = float(test_BLEU_1[0])
    elif test_BLEU_2 : 
        log_data[EPOCH]["test_BLEU_2"] = float(test_BLEU_2[0])
    elif test_BLEU_3 : 
        log_data[EPOCH]["test_BLEU_3"] = float(test_BLEU_3[0])
    elif test_BLEU_4 : 
        log_data[EPOCH]["test_BLEU_4"] = float(test_BLEU_4[0])
    elif test_METEOR : 
        log_data[EPOCH]["test_METEOR"] = float(test_METEOR[0])
    elif test_ROUGE_L : 
        log_data[EPOCH]["test_ROUGE_L"] = float(test_ROUGE_L[0])
```

```python
log_df = pd.DataFrame(log_data).T
```

```python
log_df['Epoch'] = log_df.index
```
